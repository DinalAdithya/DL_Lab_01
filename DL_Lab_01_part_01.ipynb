#**exponential**

It models the time (or distance) between events in a process where events happen continuously and independently at a constant average rate.

Common examples:


*  Time between arrivals of buses at a bus stop
*  Lifetime of a light bulb
*  Waiting time for the next customer

```python
plt.hist(np.random.rand(100000), density=True, bins=100, histtype="step", color="blue", label="Exponential")
plt.hist(np.random.rand(100000), density=True, bins=100, histtype="step", color="red", label="rand")
plt.hist(np.random.rand(100000), density=True, bins=100, histtype="step", color="green", label="randn")

plt.axis([-1, 2.5, 1, 1.1])

plt.legend(loc = "upper left")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_title("Z = X² + Y²")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.show()
```
